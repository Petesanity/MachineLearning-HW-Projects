##  Description of the dataset

In the dataset, theres data that represents the country, year, age ,gender and the suicide number for that country.
For example the first row in the dataset depicts the country of Albania in 1987 with a population of 312900 males where 21 males between the age group 15-24 commited suicide

Features:
Country: Country
Year: Year it takes place
Sex: Male/Female
Age: Age range
Suicides_no: number of suicides
Population: population of the specific gender
Suicides / 100k pop: How many people commit suicide per population
GDP_for_year: GDP of country for that year
GDP_per_capita: GDP per capita
Generation: Generation of the suicides

### Data Cleaning

```python
#copy of the dataset for cleaning
data = suicide.copy()
```

```python
#creating categorical variables for male and female
data['gender'] = data['sex'].map({"male":0, "female":1})
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
#dropping HDI
data = data.drop(["HDI for year"], axis = 1)
```

```python
#renaming columns
data=data.rename(columns={'country':'Country','year':'Year',
                          'sex': 'Sex','age':'Age','suicides_no':'SuicidesNo',
                          'population':'Population','suicides/100k pop':'Suicides100kPop',
                          'country-year':'CountryYear',
                          ' gdp_for_year ($) ':'GdpForYear','gdp_per_capita ($)':'GdpPerCapita',
                          'generation':'Generation'})
```

```python
data.info()
```

### Graphical distributions of variables


```python
#Suicide rate through the years 1985 to 2016 based on gender
#It seems males have a higher tendency to commit suicide than females
#mid 1990s suicide rate was highest

s = sns.lineplot(data = data, x = 'Year', y = 'SuicidesNo', ci = None, hue = 'Sex')
s.set_title("Suicide Numbers from 1985 to 2016 based on gender")
```

```python
#Suicide rate through the years 1985 to 2016 based on Age
#Ages 35-54 years most likely to commit suicide

s1 = sns.lineplot(data = data, x = 'Year', y = 'SuicidesNo', ci = None, hue = 'Age')
s1.set_title("#Suicide rate through the years 1985 to 2016 based on Age")
```

```python
?plt.xticks
```

```python
#Visualizing the suicide rate per population of each country
plt.figure(figsize=(20,10))
graph = sns.barplot(x="Country",y = 'Suicides100kPop', data=data, palette='colorblind')
plt.xticks(rotation=90)#rotating axis text on x-axis
graph.set_title("Suicide rate per population of each country")
```

```python
#Sorting countries by suicide rate and sorting suicide rate by valuea
s2 = pd.DataFrame(data.groupby(['Country']).Suicides100kPop.sum().reset_index()) #reset the index to the default integer index beginning at 0
s2.sort_values(by=['Suicides100kPop'],ascending=False,inplace=True)
```

```python
#Top ten countries with highest suicide rate
plt.figure(figsize = (14,8))
s4 = sns.barplot(x=s2["Country"].head(10), y=s2['Suicides100kPop'].head(10), data=s2,palette = 'colorblind')
s4.set_title("Top ten countries with highest suicide rate")
```

```python
#Check the correlation between variables via heatmap in the US
#What stands out is that there seems to be strong correlation between (population, suicideNO) in the United States

s5 = sns.heatmap(data[data['Country'] == 'United States'].corr(),annot = True)
s5.set_title("Correlation between variables via heatmap in the US")
```

```python
#Check the correlation between variables via heatmap in my home country, Jamaica
#What stands out is that there seems to be weak correlation between (population, suicideNO) in Jamaica

s7 = sns.heatmap(data[data['Country'] == 'Jamaica'].corr(),annot = True)
s7.set_title("Correlation between variables via heatmap in my home country, Jamaica")
```
